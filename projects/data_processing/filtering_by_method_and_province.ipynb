```python
# 填充 method 2 散瞳, 更精准，应该赋予更高的权重

data_2 = data[(data['验光方式1非散瞳；2散瞳；3试片']==2)]
grouped_data = data_2.groupby(['英文省份', '年份'])
for group_name, group_data in grouped_data:
    province, year = group_name
    method = 2
    value = []
    for j in range(5, 28, 3):
        group_data_0 = group_data.iloc[:, j]
        # group_data_0 = group_data.iloc[:, j].fillna('NaN')
        # group_data_1 = group_data_0.str.rstrip('%').astype(float) / 100
        mean = group_data_0.mean()
        value.append(mean)
    
    for i in range(1, 5):
        empty.loc[(empty['province'] == province) & (empty['year'] == year) & (empty['age_group'] == i), 'method']=method
        empty.loc[(empty['province'] == province) & (empty['year'] == year) & (empty['age_group'] == i), 'method_1']=method
        empty.loc[(empty['province'] == province) & (empty['year'] == year) & (empty['age_group'] == i), 'myopia_rate_urban'] = value[2*i-2]
        empty.loc[(empty['province'] == province) & (empty['year'] == year) & (empty['age_group'] == i), 'myopia_rate_rural'] = value[2*i-1]

    
empty
```

```python
# Read the estimated mean difference for calibration
# Run "Myopia_cycloplegia_diff.R" file to obtain the mean difference.

meandiff_cycloplegia_urban = pd.read_csv('../meandiff_cycloplegia_urban_4groups.csv')
meandiff_cycloplegia_rural = pd.read_csv('../meandiff_cycloplegia_rural_4groups.csv')

meansum_cycloplegia_urban = pd.read_csv('../meansum_cycloplegia_urban_4groups.csv')
meansum_cycloplegia_rural = pd.read_csv('../meansum_cycloplegia_rural_4groups.csv')
```

```python
weight_cycloplegia_urban = (2*abs(meandiff_cycloplegia_urban.iloc[:,1:])/meansum_cycloplegia_urban.iloc[:, 1:]).mean()
weight_cycloplegia_rural = (2*abs(meandiff_cycloplegia_rural.iloc[:,1:])/meansum_cycloplegia_rural.iloc[:, 1:]).mean()
weight_cycloplegia_rural = [0.95, 0.5, 0.95, 0.5] # 
# weight_cycloplegia_urban['year'] = meansum_cycloplegia_urban['year']
# weight_cycloplegia_rural['year'] = meansum_cycloplegia_rural['year']
weight_cycloplegia = (weight_cycloplegia_urban+weight_cycloplegia_rural)/2
```

```python
weight_cycloplegia_rural
```

```python
# 填充 method 1 和 3

# Fill and calibration
data_1 = data[(data['验光方式1非散瞳；2散瞳；3试片']==1) | (data['验光方式1非散瞳；2散瞳；3试片']==3)]
grouped_data = data_1.groupby(['英文省份', '年份'])
for group_name, group_data in grouped_data:
    province, year = group_name
    method = '1 or 3'
    value = []
    
    # change 100% to numbers 
    for j in range(5, 28, 3):
        group_data_0 = group_data.iloc[:, j]
        # group_data_0 = group_data.iloc[:, j].fillna('NaN')
        # group_data_1 = group_data_0.str.rstrip('%').astype(float) / 100
        mean = group_data_0.mean()
        value.append(mean)

    # Justify by the estimated difference between cycloplegia and nocycloplegia
    
    for i in range(1, 5):   
        value_orig = value[2*i-2] # orginal value to be used
        
        if empty.loc[(empty['province'] == province) & (empty['year'] == year) & (empty['age_group'] == i), 'myopia_rate_urban'].isna().any():
            empty.loc[(empty['province'] == province) & (empty['year'] == year) & (empty['age_group'] == i), 'method'] = method

            weight = weight_cycloplegia_urban.iloc[i-1]
            value_calibrate = meandiff_cycloplegia_urban.loc[(meandiff_cycloplegia_urban.year==year),"g"+str(i)].values*weight
            value_final = value_orig + value_calibrate
            empty.loc[(empty['province'] == province) & (empty['year'] == year) & (empty['age_group'] == i), 'myopia_rate_urban'] = value_final
        
        if empty.loc[(empty['province'] == province) & (empty['year'] == year) & (empty['age_group'] == i), 'myopia_rate_rural'].isna().any():
            empty.loc[(empty['province'] == province) & (empty['year'] == year) & (empty['age_group'] == i), 'method_1']=method

            weight = weight_cycloplegia_rural[i-1]
            value_calibrate = meandiff_cycloplegia_rural.loc[(meandiff_cycloplegia_rural.year==year),"g"+str(i)].values*weight
            value_final = value_orig + value_calibrate
            empty.loc[(empty['province'] == province) & (empty['year'] == year) & (empty['age_group'] == i), 'myopia_rate_rural'] = value_final
empty
```
